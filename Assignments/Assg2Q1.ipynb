Ichimoku cloud lines

```python
df['LeadingSpanA']=0
df['LeadingSpanB']=0

for i in range(51, len(df['datetime'])):

    for j in range(0,9):
        highs=[]
        lows=[]
        highs.append(float(df.iloc[i-j,[1]]))
        lows.append(float(df.iloc[i-j,[2]]))
    nineph=max(highs)
    ninepl=min(lows)
    CL= (nineph+ninepl)/2

    for j in range(0,26):
        highs=[]
        lows=[]
        highs.append(float(df.iloc[i-j,[1]]))
        lows.append(float(df.iloc[i-j,[2]]))
    tsph=max(highs)
    tspl=min(lows)
    BL= (tsph+tspl)/2

    LSA=(CL+BL)/2

    for j in range(0,52):
        highs=[]
        lows=[]
        highs.append(float(df.iloc[i-j,[1]]))
        lows.append(float(df.iloc[i-j,[2]]))
    ftph=max(highs)
    ftpl=min(lows)

    LSB=(ftph+ftpl)/2

    df["LeadingSpanA"].iloc[i]=LSA
    df["LeadingSpanB"].iloc[i]=LSB

df['LSB-LSA']=df["LeadingSpanB"]-df["LeadingSpanA"]
df
```

Ease of Movement Indicator


```python
df['EMV']=0
for i in range(1,len(df['datetime'])):
    H = float(df.iloc[i,[1]])
    L = float(df.iloc[i,[2]])
    PH = float(df.iloc[i-1,[1]] )
    PL = float(df.iloc[i-1,[2]] )
    V = float(df.iloc[i,[5]])
    EMV = (((H+L)-(PH+PL))/2)/(V/100000*(H-L))
    df["EMV"].iloc[i]=EMV

df

```

Keltner Channels


```python
#EMA = ( Vt*(2/21) ) + EMAy*(19/21)              for 20 period exponential moving average
#Opening price taken as the current value 

df['EMA']=0

for i in range(1,len(df['open'])):
    if i==1 :
        df['EMA'].iloc[i]=df['open'].iloc[i]
    else:
        df['EMA'].iloc[i]=( df['open'].iloc[i]*(2/21) ) + ( df['EMA'].iloc[i-1]*(19/21) )

df
```

```python
#Finding the Average true range
#True range is the highest of the three values found by
#1. H-L
#2. H-C.1
#3. C.1 - L
# H-Todays high
# L-Todays low
# C.1 - yesterdays close

df['TR']=0
df['ATR']=0

for i in range(1,len(df['open'])):

    tr1= float(df['high'].iloc[i])-float(df['low'].iloc[i])
    tr2= float(df['high'].iloc[i])-float(df['close'].iloc[i-1])
    tr3= float(df['close'].iloc[i-1])-float(df['low'].iloc[i])

    df['TR'].iloc[i]= max(tr1,tr2,tr3)

    sum=0
    if i<21:
        for k in (1,i+1):
            sum+=df['TR'].iloc[k]
        df['ATR'].iloc[i]=sum/i
    else:
        for j in (0,20):
            sum+=df['TR'].iloc[i-j]
        df['ATR'].iloc[i]=sum/20


df
```
