```python
import pandas as pd 
import numpy as np 
import yfinance as yf 
import warnings
warnings.filterwarnings('ignore')

df1 = yf.download('HDFCBANK.NS', start='2019-01-01', end='2023-01-01')
df2 = yf.download('RELIANCE.NS', start='2019-01-01', end='2023-01-01')
df1.reset_index(inplace=True)
df2.reset_index(inplace=True)
df1
```

```python
df2
```

```python
df1['typical price'] = (df1['High'] + df1['Low'] + df1['Close']) / 3
df1['cumulative price'] = (df1['typical price'] * df1['Volume']).cumsum()
df1['cumulative volume'] = df1['Volume'].cumsum()
df1['VWAP'] = df1['cumulative price'] / df1['cumulative volume']

df1 = df1.drop(['typical price', 'cumulative price', 'cumulative volume'], axis=1)

df1
```

```python
n = 20  #number of periods for SMA
k = 2   #number of std deviations

df1['SMA'] = df1['Close'].rolling(window=n).mean()
df1['STD'] = df1['Close'].rolling(window=n).std()

df1['Upper_Band'] = df1['SMA'] + k*df1['STD']
df1['Lower_Band'] = df1['SMA'] - k*df1['STD']

df1
```

```python
df2['typical price'] = (df2['High'] + df2['Low'] + df2['Close']) / 3
df2['cumulative price'] = (df2['typical price'] * df2['Volume']).cumsum()
df2['cumulative volume'] = df2['Volume'].cumsum()
df2['VWAP'] = df2['cumulative price'] / df2['cumulative volume']

df2 = df2.drop(['typical price', 'cumulative price', 'cumulative volume'], axis=1)

df2
```

```python
n = 20  #number of periods for SMA
k = 2   #number of std deviations
df2['SMA'] = df2['Close'].rolling(window=n).mean()
df2['STD'] = df2['Close'].rolling(window=n).std()

df2['Upper_Band'] = df2['SMA'] + k*df2['STD']
df2['Lower_Band'] = df2['SMA'] - k*df2['STD']

df2
```

```python
# entry when 
# 1. price above vwap line 
# 2. stock is oversold(price below lower band)
df1['Buy_Signal'] = np.where((df1['Close'] > df1['VWAP']) & (df1['Close'] < df1['Lower_Band']), 1, 0)

# exit when 
# 1. price below vwap line 
# 2. stock is overbought (price above upper bollinger band) 
df1['Sell_Signal'] = np.where((df1['Close'] < df1['VWAP']) | (df1['Close'] > df1['Upper_Band']), 1, 0)

df1
```

```python
#new dataframe for the buy and sell signals, storing only date and the signal
signals_df1 = pd.DataFrame(columns=['Date', 'Signal']) 

#getting buy signals from 
buy_signals1 = df1[df1['Buy_Signal'] == 1][['Date', 'Close']]       #we check for the buy signal to be 1
buy_signals1.rename(columns={'Close': 'Signal'}, inplace=True)      
buy_signals1['Signal'] = 'Buy'

#getting sell signals
sell_signals1 = df1[df1['Sell_Signal'] == 1][['Date', 'Close']]
sell_signals1.rename(columns={'Close': 'Signal'}, inplace=True)
sell_signals1['Signal'] = 'Sell'

#combining the buy and sell signal 
signals_df1 = pd.concat([buy_signals1, sell_signals1], ignore_index=True)

signals_df1['Date'] = pd.to_datetime(signals_df1['Date'])
signals_df1.sort_values('Date', inplace=True)

signals_df1
```

```python
# entry when 
# 1. price above vwap line 
# 2. stock is oversold(price below lower band)
df2['Buy_Signal'] = np.where((df2['Close'] > df2['VWAP']) & (df2['Close'] < df2['Lower_Band']), 1, 0)

# exit when 
# 1. price below vwap line 
# 2. stock is overbought (price above upper bollinger band) 
df2['Sell_Signal'] = np.where((df2['Close'] < df2['VWAP']) | (df2['Close'] > df2['Upper_Band']), 1, 0)

df2
```
